```python
#2.1  Analyse transaction trend over the years of each state
```

```python
#1

results = state_txn_users.groupby (['State', 'Year']).agg(
 total_transactions = ('Transactions', 'sum'),
 total_amount = ('Amount (INR)', 'sum')
).reset_index()

print(results.to_string(index=False))
```

```python
# 2

state_txns = state_txn_users.groupby(['State']).agg(
    total_transactions = ('Transactions','sum'),
    total_amount = ('Amount (INR)','sum')
  ).reset_index()


top_5_states_highest = state_txns.sort_values(by = 'total_transactions', ascending = False).head(5)
top_5_states_lowest = state_txns.sort_values(by = 'total_transactions', ascending = True).head(5)

print("Top 5 states with highest transactions:")
print(top_5_states_highest)

print("Top 5 states with lowest transactions:")
print(top_5_states_lowest)
```

```python
# 4.1 Ratio of users to population by state
```

```python
#1 Merge the state_txn_users dataset with the district_demographics dataset to calculate the ratio of registered users to the population of each state
#display the results in a tabular format.


state_population_df  = district_demographics.groupby('State').agg(
    Population = ('Population','sum')
).reset_index()

state_registered_users = state_txn_users.groupby('State').agg(
    Registered_users = ('Registered Users','sum')
).reset_index()

merge_dfs = pd.merge(state_population_df, state_registered_users, on='State', how='inner')

merge_dfs['user_to_population_ratio'] = merge_dfs['Registered_users']/merge_dfs['Population']

print(merge_dfs.to_string(index=False))
```

```python
#2  Create a column chart depicting the ratio of users to population by state
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
plt.bar(merge_dfs['State'], merge_dfs['user_to_population_ratio'], color = 'orange')

plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('user_to_population_ratio')
plt.title('State wise user to population ratio')
plt.grid(axis='y', label = '--')
plt.tight_layout()

plt.show()
```

```python
# 4.2 Correlate population density with transaction volume
```

```python
#4.4 Device brand usage ratio
```

```python
#1 Merge the state_device_data dataset with the state_txn_users dataset.

merge_data = pd.merge(state_device_data, state_txn_users, on = ['State','Year','Quarter'], how = 'inner')
print(merge_data)
```

```python
#2 Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the result tabular format

grouped_data = merge_data.groupby(["State","Brand"], as_index=False).agg({
    'Registered Users_x':'sum',
    'Registered Users_y':'max'
})

grouped_data['User_ratio'] = grouped_data['Registered Users_x']/grouped_data['Registered Users_y']

grouped_data['User_ratio (%)'] = grouped_data['User_ratio']* 100

ratio_results = grouped_data[['State','Brand','Registered Users_x','Registered Users_y','User_ratio','User_ratio (%)']]
ratio_results = ratio_results.sort_values(['State','User_ratio'], ascending = [True,False])
```

```python
print(ratio_results.to_string(index=False))
```

```python
#3 Create a bar chart depicting the device brand usage ratio for each state.

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (16,8))

sns.barplot(
    data = ratio_results,
    x = 'State',
    y = 'User_ratio',
    hue = 'Brand'
)

plt.xticks(rotation = 90)

plt.legend(title = 'Brand',loc = 'best')

plt.xlabel('State')
plt.ylabel('User ratio')
plt.title('State wise device brand usage ratio')

plt.grid(axis = 'y', label = '--')
plt.tight_layout()
plt.show()
```
